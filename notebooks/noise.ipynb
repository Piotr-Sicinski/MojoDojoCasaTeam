```python
from qiskit import QuantumCircuit, transpile
from qiskit.providers.aer import AerSimulator
from qiskit.visualization import plot_histogram
from math import asin, sqrt

# Create a quantum circuit with 4 qubits
qc = QuantumCircuit(81)

# Rotation angles for 90% and 10% probabilities
theta_90 = 2 * asin(sqrt(0.95))
theta_10 = 2 * asin(sqrt(0.05))

# Apply Ry-gate with theta_10 on qubits 0 and 2
qc.ry(theta_10, 0)
qc.ry(theta_10, 2)

# Apply Ry-gate with theta_90 on qubits 1 and 3
qc.ry(theta_90, 1)
qc.ry(theta_90, 3)

# Measure the qubits
qc.measure_all()

# Simulate the circuit
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
result = simulator.run(compiled_circuit).result()

# Get and display the histogram of the results
counts = result.get_counts(qc)
plot_histogram(counts)


```
